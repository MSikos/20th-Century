## 1. Install libraries

```python
from textblob import TextBlob 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import nltk
import re
import os
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
from collections import Counter
sns.set()
```

## 2. Load the 20th Century Events text file and tokenize it

```python
# Import txt file
with open('20thCentury_Events_Wiki.txt', 'r', errors='ignore') as file:
    data = file.read().replace('\n', '')
```

```python
# Replace United States and United Kingdom from 2 tag words to one tag
data = data.replace("United States", "USA")
data = data.replace("U.S", "USA")
data = data.replace("USA", "USA")
data = data.replace("United Kingdom", "U.K")                   
```

```python
# Word tokenization
from nltk.tokenize import word_tokenize
tokenized_word = word_tokenize(data)
print(tokenized_word)
```

## 3. Create a frequency distribution of the most common words

```python
# Create frequency distribution
from nltk.probability import FreqDist
dist_words = FreqDist(tokenized_word)
print(dist_words)
```

```python
dist_words.most_common(10)
```

```python
# Frequency Distribution Plot
plt.figure(figsize=(8, 3))
dist_words.plot(10,cumulative = False)
plt.show()
```

## 4. Create Stopwords

```python
# Defining stopwords
from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))
print(stop_words)
```

```python
# Removing stopwords in words
filtered_words = [] # creates an empty list
for word in tokenized_word:
    if word not in stop_words:
        filtered_words.append(word)
```

```python
filtered_words
```

```python
# Create a new FreqDist for filtered_words
dist_words_filter = FreqDist(filtered_words)
print(dist_words_filter)
```

```python
# Frequency Distribution Plot
plt.figure(figsize=(8, 3))
dist_words_filter.plot(10, cumulative = False)
plt.show()
```

```python
dist_words_filter
```

```python
#Substitute all punctuations marks with a space 
sans_punc = re.sub("[^a-zA-Z]",  # Search for all non-letters
                          " ",          # Replace all non-letters with spaces
                          str(filtered_words))
sans_punc
```

```python
# Word tokenization
tokenized_word_2 = word_tokenize(sans_punc)
print(tokenized_word_2)
```

```python
# Create a new FreqDist
dist_words_filter_2 = FreqDist(tokenized_word_2)
```

```python
# Frequency Distribution Plot
plt.figure(figsize=(8, 3))
dist_words_filter_2.plot(30, cumulative = False)
plt.show()
```

## <b>Impressions:

### Prior to removing the stopwords and punctuation, the majority of words used was punctuation like commas and periods and the word The. The word "The" will be removed in the next step whe I create the POS tags with TextBlob

## 5. POS tags with TextBlob

```python
# Create a new Stopwords list
new_stopwords = ["And", "Then", 'n', 't', 's', 'th', 'The', 'www', 'ISBN', 'edit']
```

```python
filtered = []
for word in tokenized_word_2:
    if word not in new_stopwords:
        filtered.append(word)
```

```python
%%time
text = TextBlob(str(filtered))
```

```python
text
```

```python
# Create a list of the POS tags in the text file
tags_list = text.tags
tags_list
```

```python
# Create a dataframe from the list of POS tags
df_text = pd.DataFrame(tags_list)
df_text.columns = ['Words', "Word type"]
df_text.head()
```

```python
# Grouping the POS tags by type
df_t = df_text.groupby('Word type').count().reset_index()
df_t
```

```python
# Create a dataframe with the top 10 POS tags
top10 = df_t.nlargest(10, 'Words')
```

```python
# Create a bar plot of the top 10 POS tags
plt.figure(figsize = (10, 5))
with sns.dark_palette("xkcd:blue", 20):
    sns.barplot(x = "Words", y = "Word type",
    saturation = 0.9, data = top10).set_title("Key Events in the 20th Century - top 10 word types used")
```

## 6. Top 15 POS for Nouns

```python
df = df_text[(df_text['Word type'] == "NN") | (df_text['Word type'] == "NNS") | (df_text['Word type'] == "NNP")]
df.columns = ["Word", "Occurences"]
x = df.groupby('Word').count().reset_index()
y = x.sort_values(by = ['Occurences'], ascending=False)
top15 = y.nlargest(15, 'Occurences')
top15
```

```python
# Create a barchart of the 15 most used nouns
plt.figure(figsize=(15, 5))
with sns.dark_palette("xkcd:blue", 15):
    sns.barplot(x="Word", y="Occurences",
    saturation=0.9, data = top15).set_title("Key Events in the 20th Centuyry - most frequently used nouns")
```

## 7. Top 15 POS for Verbs

```python
# Create a dataframe with the 15 most commonly used Verbs
df = df_text[(df_text['Word type'] == "VB")  | (df_text['Word type'] == "VBD")]
df.columns = ["Word", "Occurences"]
x = df.groupby('Word').count().reset_index()
y = x.sort_values(by = ['Occurences'], ascending=False)
top15 = y.nlargest(15, 'Occurences')
top15
```

```python
# Create a bar plot of the 15 most commonly used Verbs
plt.figure(figsize = (15, 5))
with sns.dark_palette("xkcd:blue", 15):
    sns.barplot(x = "Word", y = "Occurences",
    saturation = 0.9, data = top15).set_title("Key Events of the 20th Century - most frequently used verbs")
```

## 8. Top 15 POS for Adjectives

```python
# Create a dataframe with the 15 most commonly used Adjectives
df = df_text[df_text['Word type'] == "JJ"]
df.columns = ["Word", "Occurences"]
x = df.groupby('Word').count().reset_index()
y = x.sort_values(by=['Occurences'], ascending=False)
top15 = y.nlargest(15, 'Occurences')
top15
```

```python
# Create a bar plot of the 15 most commonly used Adjectives
plt.figure(figsize=(15, 5))
with sns.dark_palette("xkcd:blue", 15):
    sns.barplot(x="Word", y="Occurences",
    saturation=0.9, data=top15).set_title("Key Events of the 20th Century - most frequently used adjectives")
```

### <b>Impressions:

Looking at the box plots, several patterns appear in the noun, verb, and adjective plots. The majority of the words relate or refer to World War 2 or events centered around World War 2. Looking at the Verb plot, I would also add the word "Retrieved" to the Stopwords list, since that refers to when the source data was obtained to write or edit the Wikipedia page and does not carry any significance to the analysis.

## 10. Bonus Task

Given that the web page that was scraped was from Wikipedia, I would expect the polarity and subjectivity to be close to neutral polarity and subjectivity as possible. Wikipedia tends to call out bias and ask for or require revision since its mission is to be a free, open-source encyclopedia that reports the facts objectively.

```python
# Import Sentiment Intensity Analyzer to analyze sentences for sentiment
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
# Download NlTK Data
nltk.download('vader_lexicon')
```

```python
# Import txt file
with open('20thCentury_Events_Wiki.txt', 'r', errors='ignore') as file:
    text = file.read()
```

```python
# Perform Sentiment Analysis
sid = SentimentIntensityAnalyzer()
sentences = nltk.sent_tokenize(text)
sentiment_scores = [sid.polarity_scores(sentence) for sentence in sentences]
```

```python
# Extract Sentiment scores
positive = [score['pos'] for score in sentiment_scores]
negative = [score['neg'] for score in sentiment_scores]
neutral = [score['neu'] for score in sentiment_scores]
```

```python
# Extract compound scores
compound_scores = [score['compound'] for score in sentiment_scores]
```

```python
# Plot sentiment scores
plt.figure(figsize=(10, 6))
plt.plot(compound_scores, marker='o')
plt.title('Sentiment Analysis of Key Events in the 20th Century')
plt.xlabel('Sentence Number')
plt.ylabel('Sentiment Score')
plt.grid(True)
plt.show()
```
